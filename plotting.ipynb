```python
import re
import datetime
import pandas
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from parsing_whatsapp import read_messages
```

```python
df=read_messages('chat.txt')
```

```python
reply_duration=[]
for i in range (len(df)-1):
    if (df['author'][i]!=df['author'][i+1] and df['author'][i]=='ameesh bits cs'):
        reply_duration.append((df['time'][i+1]-df['time'][i]).seconds/60.0)
```

```python
for i in range(len(reply_duration)-1):
    if reply_duration[i]==0:
        del reply_duration[i]

print(reply_duration)
```

```python
plt.bar(range(len(reply_duration)),reply_duration)
plt.title("Sender's Reply time in Minutes vs Time")
plt.text(10,1250, 'Average time taken to reply:  166.8641975308642 minutes\nStd. Deviation:  331.97035242777736 minutes', fontsize =7)
plt.show()
```

```python
plt.plot(reply_duration)
plt.show()
```

```python
reply_duration=np.array(reply_duration)
print("Average time taken to reply: ",np.mean(reply_duration),"minutes\nStd. Deviation: ", np.std(reply_duration), "minutes")
```
